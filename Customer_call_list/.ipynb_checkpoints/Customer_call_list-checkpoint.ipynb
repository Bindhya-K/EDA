```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('Customer Call List.xlsx')
df.head()
```

```python
df.info()
```

### 1.  removing duplicates if any

```python
df = df.drop_duplicates()
```

```python
df = df.drop(['Not_Useful_Column'],axis=1)
```

### 2. Last name column has some characters

```python
df['Last_Name']= df['Last_Name'].str.extract('([a-z|A-Z]+)',expand=True)
```

### 3. Making phone number in the format "123-456-789 "

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
```

```python
df['Phone_Number']=df['Phone_Number'].astype('string')
```

```python
df.Phone_Number.dtype
```

```python
df['Phone_Number']= df.Phone_Number.fillna('000000000')
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('000-000-000','').str.replace('Na--','')
```

```python
df
```

###  4. Address column splitting 

```python
df[['Street_Name','City','Zip']]=df['Address'].str.split(',',expand=True)
```

```python
df.drop('Address',axis=1,inplace=True)
```

### 5. Paying Customer- Y as Yes and N as No 

```python
#df['Paying Customer'] = df['Paying Customer'].str.replace('(Y|y)','Yes')
```

```python
a= df['Paying Customer'].str.replace('Yes','Y').str.replace('No','N')
```

```python
df['Paying Customer'] = a.str.replace('Y','Yes').str.replace('N','No')
```

```python
df['Paying Customer'] =df['Paying Customer'] .str.replace('No/a','')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace("No",'N').str.replace('Yes','Y')
df
```


### 6. removing curtomers who do npt hav phone numbers and y as do not contact

```python
for x in df.index:
    if df.loc[x,'Do_Not_Contact'] == 'Y':
        df.drop(x,inplace= True)
```

```python
df.dropna(subset=['Do_Not_Contact'],inplace=True)
```

```python
for x in df.index:
    if df.loc[x,'Phone_Number'] == '':
        df.drop(x,inplace= True)
```
