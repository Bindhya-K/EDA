```python
df.dropna()
```

```python
df1 = pd.read_excel('Country_code.xlsx')
```

```python
df1.head()
```

```python
# concatinating 2 dataframes
final_df = pd.merge(df,df1,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

### 2. Exploring data columns

```python
final_df.Country.value_counts()
```

Insight: More transaction in India, then USA, then UK .....

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
# pie chart
plt.pie(country_val, labels=country_names)
```

```python
# to get a more clear view, plot top 3 countries
# pie chart
plt.pie(country_val[0:3], labels=country_names[:3])
```

```python
# pie chart to plat values also
plt.pie(country_val[:3], labels=country_names[:3],autopct='%1.1f%%')
```

```python
# rating from different countries:
final_df['Aggregate rating'].values
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating_Counts'})
```

```python
ratings
```

#### raitng colour: white ==> not rated
#### rating colour: Red ==> Poor (1-2.4 )
#### rating colour: Orange ==> Average(2.5-3.4)
#### rating colour: Yellow==> Good(3.5-3.9)
#### rating colour: Green ==> Very Good (4-4.4)
#### rating colour: Dark Green ==> Excellent (4.5 - 4.9)

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(ratings,x='Aggregate rating',y='Rating_Counts')
```

```python
final_df.columns
```

```python
# to change the colours of barplot w.r.t aggregate colours
sns.barplot(ratings,x='Aggregate rating',y='Rating_Counts',hue='Rating color')
```

```python
# but still colour is not mapping
sns.barplot(ratings,x='Aggregate rating',y='Rating_Counts',hue='Rating color',palette=['Blue','Red','Orange','Yellow','Green','Green'])
```

Observation :
1. Not rated count - high
2. max no. of ratings - 2.8 to 4

```python
# count plot to plot categorical values
sns.countplot(ratings,x='Rating color',palette=['Blue','Red','Orange','Yellow','Green','Green'])
```
