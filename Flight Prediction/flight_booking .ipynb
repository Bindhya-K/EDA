```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
test_df=pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df=train_df.append(test_df)
final_df.head()
```

```python
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head(10)
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
# or final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
final_df['Arrival_hr']=final_df['Arrival_Time'].apply(lambda x: x.split(':')[0]).astype(int)
final_df['Arrival_min']=final_df['Arrival_Time'].apply(lambda x: x.split(':')[1]).astype(int)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df['Dep_time_hr']=final_df['Dep_Time'].apply(lambda x: x.split(':')[0]).astype(int)
final_df['Dep_time_min']=final_df['Dep_Time'].apply(lambda x: x.split(':')[1]).astype(int)
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df.head()
```

```python
final_df['Duration_hr'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df['Duration_hr'].unique()
```

```python
final_df[final_df['Duration_hr']=='5m']
```

```python
# 5m row can be discarde
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_hr'] = final_df['Duration_hr'].astype(int)
```

```python
final_df['Duration_hr'] = final_df['Duration_hr']*60
```

```python
final_df['Duration_min'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df['Duration_min'].fillna('0',inplace=True)
```

```python
final_df['Duration_min'] = final_df['Duration_min'].astype(float)
```

```python
final_df['Duration_min']  = final_df['Duration_min'] *(1/60)
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```

```python
final_df['Duration']=final_df['Duration_hr'] + final_df['Duration_min'] 
```

```python
final_df.drop(['Duration_hr','Duration_min'],axis=1,inplace=True)
```
